```python
#import important libraries 
import pandas as pd
import numpy as np
```

```python
#reading json files and then turning them to dataframes
df1=pd.read_json('12387.json', orient='index')
df1
```

```python
df2=pd.read_json('16859.json', orient='index')
df3=pd.read_json('18246.json', orient='index')
```

```python
dataframe1=pd.DataFrame(df1)
dataframe2=pd.DataFrame(df2)
dataframe3=pd.DataFrame(df3)
```

```python
#we need to transpose dataframes before concating them
dataframe1 =dataframe1.transpose()
dataframe2 =dataframe2.transpose()
dataframe3 =dataframe3.transpose()
```

```python
data=[dataframe1,dataframe2,dataframe3]
DataFrame=pd.concat(data)
```

```python
#droping the columns we do not need
New_DF=DataFrame[['id','text','words','audio_url','utterances','audio_duration','audio_url']]

New_DF
```

```python
#get the text column
text=New_DF['text']

```

```python
###get our dataset
Frame=pd.read_csv('to_fill.csv')
Frame
```

```python
d0=pd.DataFrame(Frame)
```

```python
#### finding body , start, and end time of story

first_words=d0['first_words']
body = []
start_time = []
end_time = []
#looping through first words and the texts
for line in first_words:
  for passage in text:

    #checking if first line is in the passage
    if line in passage:

      #get starting index of the line
      start=passage.find(line)

      #get end line from dataset
      endline=d0['last_words'][d0['first_words']==line]
      
      # driving body from passage using the first and last indexs 
      for txt in endline:
        point=passage.find(txt)
        endpoint=passage.rfind(txt)
        story=passage[start:endpoint] +txt
        body.append(story)
        print(f" body passage is:{story}")

      ####finding the start and end time of story
      #get the words information from dataset for related passage
      words_target=New_DF['words'][New_DF['text']==passage]
      count_word=passage.split()
      count_body=story.split()
      count_to_end=passage[start:].split()
      
      num_first=len(count_word) - len(count_to_end)
      num_last= num_first + len(count_body)
      
      
      print(f" words information : {words_target}")
      #looping through list of dictionaries to find the start and end time of the story
      for i in words_target:
        start_word=i[num_first]
        start_time.append(start_word['start'])
        #we have some missing data in words so we have to check if number of dictionaries are appropriate 
        if num_last-1<len(count_word) :
          end_word= i[num_last-1]
          end_time.append(end_word['end'])
          print(f", last word:{end_word['text']}")
        else:
          end_time.append(0)
        
        print(f"start time:{start_word['text']},")


      print('/////////////')


      
```

```python

body_col= pd.DataFrame(body, columns=['body'])
body_col
```

```python
start_col=pd.DataFrame(start_time, columns=['strat'])
start_col
```

```python
end_col=pd.DataFrame(end_time, columns=['end'])
end_col
```

```python
d0['body']=body_col
d0['start']=start_col
d0['end']=end_col
d0
```

```python
# Define the filename for the CSV file
filename = 'news.csv'

# Save the DataFrame as a CSV file
d0.to_csv(filename, index=False)

```
